```
%%sql -r df_policy
-- Create POLICY table (40 rows) with FK to INSURED and LOB
CREATE OR REPLACE TABLE INSURANCE_CLAIM_DB.CLAIMS_SCHEMA.POLICY (
    POLICY_ID INT PRIMARY KEY,
    POLICY_NUMBER VARCHAR(20),
    INSURED_ID INT,
    LOB_ID INT,
    POLICY_EFFECTIVE_DATE DATE,
    POLICY_EXPIRATION_DATE DATE,
    POLICY_TERM_MONTHS INT,
    COVERAGE_TYPE VARCHAR(50),
    COVERAGE_LIMIT DECIMAL(15,2),
    DEDUCTIBLE DECIMAL(10,2),
    PREMIUM DECIMAL(10,2),
    POLICY_STATE VARCHAR(2),
    POLICY_ZIP VARCHAR(10),
    AGENT_NAME VARCHAR(100),
    AGENCY_NAME VARCHAR(100),
    POLICY_STATUS VARCHAR(20),
    FOREIGN KEY (INSURED_ID) REFERENCES INSURANCE_CLAIM_DB.CLAIMS_SCHEMA.INSURED(INSURED_ID),
    FOREIGN KEY (LOB_ID) REFERENCES INSURANCE_CLAIM_DB.CLAIMS_SCHEMA.LINE_OF_BUSINESS(LOB_ID)
);

INSERT INTO INSURANCE_CLAIM_DB.CLAIMS_SCHEMA.POLICY
SELECT 
    SEQ4() + 1 AS POLICY_ID,
    'POL-' || LPAD((SEQ4() + 1)::VARCHAR, 6, '0') AS POLICY_NUMBER,
    (SEQ4() * 200) + 1 AS INSURED_ID,
    MOD(SEQ4(), 15) + 1 AS LOB_ID,
    DATEADD(DAY, -MOD(SEQ4() * 47, 1800), '2025-01-01'::DATE) AS POLICY_EFFECTIVE_DATE,
    DATEADD(MONTH, CASE MOD(SEQ4(), 3) WHEN 0 THEN 6 WHEN 1 THEN 12 ELSE 24 END, 
        DATEADD(DAY, -MOD(SEQ4() * 47, 1800), '2025-01-01'::DATE)) AS POLICY_EXPIRATION_DATE,
    CASE MOD(SEQ4(), 3) WHEN 0 THEN 6 WHEN 1 THEN 12 ELSE 24 END AS POLICY_TERM_MONTHS,
    CASE MOD(SEQ4(), 5) WHEN 0 THEN 'Comprehensive' WHEN 1 THEN 'Liability Only' WHEN 2 THEN 'Collision' 
        WHEN 3 THEN 'Full Coverage' ELSE 'Basic' END AS COVERAGE_TYPE,
    CASE MOD(SEQ4(), 4) WHEN 0 THEN 100000 WHEN 1 THEN 250000 WHEN 2 THEN 500000 ELSE 1000000 END AS COVERAGE_LIMIT,
    CASE MOD(SEQ4(), 4) WHEN 0 THEN 500 WHEN 1 THEN 1000 WHEN 2 THEN 2500 ELSE 5000 END AS DEDUCTIBLE,
    ROUND(500 + MOD(SEQ4() * 113, 4500), 2) AS PREMIUM,
    CASE MOD(SEQ4(), 10) WHEN 0 THEN 'CA' WHEN 1 THEN 'TX' WHEN 2 THEN 'FL' WHEN 3 THEN 'NY' WHEN 4 THEN 'PA'
        WHEN 5 THEN 'IL' WHEN 6 THEN 'OH' WHEN 7 THEN 'GA' WHEN 8 THEN 'NC' ELSE 'MI' END AS POLICY_STATE,
    LPAD((10000 + MOD(SEQ4() * 17, 90000))::VARCHAR, 5, '0') AS POLICY_ZIP,
    'Agent_' || (MOD(SEQ4(), 50) + 1)::VARCHAR AS AGENT_NAME,
    CASE MOD(SEQ4(), 5) WHEN 0 THEN 'Premier Insurance Agency' WHEN 1 THEN 'Trusted Coverage Group' 
        WHEN 2 THEN 'AllState Partners' WHEN 3 THEN 'SecureLife Brokers' ELSE 'National Insurance Services' END AS AGENCY_NAME,
    CASE WHEN MOD(SEQ4(), 7) = 0 THEN 'Cancelled' ELSE 'Active' END AS POLICY_STATUS
FROM TABLE(GENERATOR(ROWCOUNT => 40));
```

```
%%sql -r df_claims_insert
-- Insert 10000 Claims with proper FK references, fraud markers, severity changes
INSERT INTO INSURANCE_CLAIM_DB.CLAIMS_SCHEMA.CLAIMS
SELECT 
    SEQ4() + 1 AS CLAIM_ID,
    'CLM-' || LPAD((SEQ4() + 1)::VARCHAR, 8, '0') AS CLAIM_NUMBER,
    MOD(SEQ4(), 40) + 1 AS POLICY_ID,
    MOD(SEQ4(), 8000) + 1 AS INSURED_ID,
    MOD(SEQ4(), 15) + 1 AS LOB_ID,
    DATEADD(DAY, MOD(SEQ4() * 23, 2262), '2020-01-01'::DATE) AS LOSS_DATE,
    DATEADD(DAY, MOD(SEQ4() * 23, 2262) + MOD(SEQ4(), 14) + 1, '2020-01-01'::DATE) AS REPORTED_DATE,
    DATEADD(DAY, MOD(SEQ4() * 23, 2262) + MOD(SEQ4(), 14) + MOD(SEQ4(), 7) + 2, '2020-01-01'::DATE) AS OPEN_DATE,
    CASE WHEN MOD(SEQ4(), 100) < 60 
         THEN DATEADD(DAY, MOD(SEQ4() * 23, 2262) + MOD(SEQ4() * 7, 180) + 30, '2020-01-01'::DATE) 
         ELSE NULL END AS CLOSE_DATE,
    CASE MOD(SEQ4(), 15)
        WHEN 0 THEN 'Vehicle Collision' WHEN 1 THEN 'Property Theft' WHEN 2 THEN 'Water Damage' 
        WHEN 3 THEN 'Fire Damage' WHEN 4 THEN 'Slip and Fall' WHEN 5 THEN 'Medical Injury'
        WHEN 6 THEN 'Wind Damage' WHEN 7 THEN 'Vandalism' WHEN 8 THEN 'Equipment Failure'
        WHEN 9 THEN 'Natural Disaster' WHEN 10 THEN 'Workplace Injury' WHEN 11 THEN 'Product Defect'
        WHEN 12 THEN 'Professional Error' WHEN 13 THEN 'Cyber Attack' ELSE 'Hail Damage'
    END AS LOSS_CAUSE,
    'Claim for ' || CASE MOD(SEQ4(), 15)
        WHEN 0 THEN 'vehicle collision damage' WHEN 1 THEN 'property theft incident' WHEN 2 THEN 'water damage to property'
        ELSE 'reported incident' END || ' - Claim #' || (SEQ4() + 1)::VARCHAR AS LOSS_DESCRIPTION,
    CASE WHEN MOD(SEQ4(), 100) < 15 THEN TRUE ELSE FALSE END AS CAT_INDICATOR,
    CASE WHEN MOD(SEQ4(), 100) < 15 THEN MOD(SEQ4(), 15) + 1 ELSE NULL END AS CATASTROPHE_ID,
    CASE WHEN MOD(SEQ4(), 100) < 12 THEN TRUE ELSE FALSE END AS LITIGATION_INDICATOR,
    CASE MOD(SEQ4(), 5) WHEN 0 THEN 'Minor' WHEN 1 THEN 'Moderate' WHEN 2 THEN 'Significant' 
        WHEN 3 THEN 'Severe' ELSE 'Catastrophic' END AS CLAIM_SEVERITY,
    CASE WHEN MOD(SEQ4(), 4) = 0 THEN 
        CASE MOD(SEQ4() + 1, 5) WHEN 0 THEN 'Minor' WHEN 1 THEN 'Moderate' WHEN 2 THEN 'Significant' 
            WHEN 3 THEN 'Severe' ELSE 'Catastrophic' END 
        ELSE NULL END AS PREVIOUS_SEVERITY,
    CASE WHEN MOD(SEQ4(), 4) = 0 THEN DATEADD(DAY, -MOD(SEQ4(), 60), CURRENT_DATE()) ELSE NULL END AS SEVERITY_CHANGE_DATE,
    CASE WHEN MOD(SEQ4(), 4) = 0 THEN 
        CASE MOD(SEQ4(), 6) WHEN 0 THEN 'Initial Report' WHEN 1 THEN 'Under Investigation' 
            WHEN 2 THEN 'Documentation Review' WHEN 3 THEN 'Adjuster Assigned'
            WHEN 4 THEN 'Damage Assessment' ELSE 'Liability Determination' END 
        ELSE NULL END AS SEVERITY_CHANGE_STAGE,
    CASE MOD(SEQ4(), 100)
        WHEN 0 THEN 'Closed' WHEN 1 THEN 'Closed' WHEN 2 THEN 'Closed' WHEN 3 THEN 'Closed' WHEN 4 THEN 'Closed'
        WHEN 5 THEN 'Closed' WHEN 6 THEN 'Closed' WHEN 7 THEN 'Closed' WHEN 8 THEN 'Closed' WHEN 9 THEN 'Closed'
        WHEN 10 THEN 'Closed' WHEN 11 THEN 'Closed' WHEN 12 THEN 'Closed' WHEN 13 THEN 'Closed' WHEN 14 THEN 'Closed'
        WHEN 15 THEN 'Closed' WHEN 16 THEN 'Closed' WHEN 17 THEN 'Closed' WHEN 18 THEN 'Closed' WHEN 19 THEN 'Closed'
        WHEN 20 THEN 'Closed' WHEN 21 THEN 'Closed' WHEN 22 THEN 'Closed' WHEN 23 THEN 'Closed' WHEN 24 THEN 'Closed'
        WHEN 25 THEN 'Closed' WHEN 26 THEN 'Closed' WHEN 27 THEN 'Closed' WHEN 28 THEN 'Closed' WHEN 29 THEN 'Closed'
        WHEN 30 THEN 'Closed' WHEN 31 THEN 'Closed' WHEN 32 THEN 'Closed' WHEN 33 THEN 'Closed' WHEN 34 THEN 'Closed'
        WHEN 35 THEN 'Closed' WHEN 36 THEN 'Closed' WHEN 37 THEN 'Closed' WHEN 38 THEN 'Closed' WHEN 39 THEN 'Closed'
        WHEN 40 THEN 'Closed' WHEN 41 THEN 'Closed' WHEN 42 THEN 'Closed' WHEN 43 THEN 'Closed' WHEN 44 THEN 'Closed'
        WHEN 45 THEN 'Approved' WHEN 46 THEN 'Approved' WHEN 47 THEN 'Approved' WHEN 48 THEN 'Approved' WHEN 49 THEN 'Approved'
        WHEN 50 THEN 'Approved' WHEN 51 THEN 'Approved' WHEN 52 THEN 'Approved' WHEN 53 THEN 'Approved' WHEN 54 THEN 'Approved'
        WHEN 55 THEN 'Rejected' WHEN 56 THEN 'Rejected' WHEN 57 THEN 'Rejected' WHEN 58 THEN 'Rejected' WHEN 59 THEN 'Rejected'
        WHEN 60 THEN 'Rejected' WHEN 61 THEN 'Rejected' WHEN 62 THEN 'Rejected' WHEN 63 THEN 'Rejected' WHEN 64 THEN 'Rejected'
        WHEN 65 THEN 'Open' WHEN 66 THEN 'Open' WHEN 67 THEN 'Open' WHEN 68 THEN 'Open' WHEN 69 THEN 'Open'
        WHEN 70 THEN 'Open' WHEN 71 THEN 'Open' WHEN 72 THEN 'Open' WHEN 73 THEN 'Open' WHEN 74 THEN 'Open'
        WHEN 75 THEN 'Stalled' WHEN 76 THEN 'Stalled' WHEN 77 THEN 'Stalled' WHEN 78 THEN 'Stalled' WHEN 79 THEN 'Stalled'
        WHEN 80 THEN 'Stalled' WHEN 81 THEN 'Stalled' WHEN 82 THEN 'Stalled' WHEN 83 THEN 'Stalled' WHEN 84 THEN 'Stalled'
        WHEN 85 THEN 'Stalled' WHEN 86 THEN 'Stalled' WHEN 87 THEN 'Stalled' WHEN 88 THEN 'Stalled' WHEN 89 THEN 'Stalled'
        ELSE 'Pending Review'
    END AS CLAIM_STATUS,
    CASE MOD(SEQ4(), 12)
        WHEN 0 THEN 'Initial Report' WHEN 1 THEN 'Under Investigation' WHEN 2 THEN 'Documentation Review'
        WHEN 3 THEN 'Adjuster Assigned' WHEN 4 THEN 'Damage Assessment' WHEN 5 THEN 'Liability Determination'
        WHEN 6 THEN 'Negotiation' WHEN 7 THEN 'Settlement Review' WHEN 8 THEN 'Payment Processing'
        WHEN 9 THEN 'Closed' WHEN 10 THEN 'Reopened' ELSE 'Litigation'
    END AS CLAIM_STAGE,
    CASE MOD(SEQ4(), 10) WHEN 0 THEN 'CA' WHEN 1 THEN 'TX' WHEN 2 THEN 'FL' WHEN 3 THEN 'NY' WHEN 4 THEN 'PA'
        WHEN 5 THEN 'IL' WHEN 6 THEN 'OH' WHEN 7 THEN 'GA' WHEN 8 THEN 'NC' ELSE 'MI' END AS CLAIM_STATE,
    CASE MOD(SEQ4(), 24)
        WHEN 0 THEN 'Los Angeles' WHEN 1 THEN 'San Diego' WHEN 2 THEN 'San Francisco' WHEN 3 THEN 'Houston' 
        WHEN 4 THEN 'Dallas' WHEN 5 THEN 'Austin' WHEN 6 THEN 'Miami' WHEN 7 THEN 'Tampa' WHEN 8 THEN 'Orlando'
        WHEN 9 THEN 'New York' WHEN 10 THEN 'Buffalo' WHEN 11 THEN 'Albany' WHEN 12 THEN 'Philadelphia' 
        WHEN 13 THEN 'Pittsburgh' WHEN 14 THEN 'Chicago' WHEN 15 THEN 'Aurora' WHEN 16 THEN 'Columbus' 
        WHEN 17 THEN 'Cleveland' WHEN 18 THEN 'Atlanta' WHEN 19 THEN 'Savannah' WHEN 20 THEN 'Charlotte' 
        WHEN 21 THEN 'Raleigh' WHEN 22 THEN 'Detroit' ELSE 'Grand Rapids'
    END AS CLAIM_CITY,
    LPAD((10000 + MOD(SEQ4() * 13, 90000))::VARCHAR, 5, '0') AS CLAIM_ZIP,
    CASE WHEN MOD(SEQ4(), 100) < 8 THEN TRUE ELSE FALSE END AS FRAUD_INDICATOR,
    CASE WHEN MOD(SEQ4(), 100) < 8 THEN 50 + MOD(SEQ4() * 7, 50) 
         WHEN MOD(SEQ4(), 100) < 20 THEN MOD(SEQ4() * 3, 50) ELSE 0 END AS FRAUD_SCORE,
    CASE WHEN MOD(SEQ4(), 100) < 18 THEN TRUE ELSE FALSE END AS SUBROGATION_INDICATOR,
    MOD(SEQ4(), 200) + 1 AS VENDOR_ID,
    MOD(SEQ4(), 5) + 1 AS EXPOSURES,
    DATEADD(DAY, -MOD(SEQ4(), 90), CURRENT_DATE()) AS LAST_ACTIVITY_DATE,
    DATEADD(DAY, MOD(SEQ4() * 23, 2262) + MOD(SEQ4(), 7), '2020-01-01'::DATE) AS CREATED_DATE,
    DATEADD(DAY, -MOD(SEQ4(), 30), CURRENT_DATE()) AS MODIFIED_DATE
FROM TABLE(GENERATOR(ROWCOUNT => 10000));
```

```
%%sql -r df_demo_q2
-- Demo Q2: What is the % of approved Vs rejected claims by geography?
SELECT 
    CLAIM_STATE AS GEOGRAPHY,
    COUNT(*) AS TOTAL_CLAIMS,
    COUNT(CASE WHEN CLAIM_STATUS = 'Approved' THEN 1 END) AS APPROVED_COUNT,
    COUNT(CASE WHEN CLAIM_STATUS = 'Rejected' THEN 1 END) AS REJECTED_COUNT,
    ROUND(COUNT(CASE WHEN CLAIM_STATUS = 'Approved' THEN 1 END) * 100.0 / NULLIF(COUNT(*), 0), 2) AS APPROVED_PCT,
    ROUND(COUNT(CASE WHEN CLAIM_STATUS = 'Rejected' THEN 1 END) * 100.0 / NULLIF(COUNT(*), 0), 2) AS REJECTED_PCT
FROM INSURANCE_CLAIM_DB.CLAIMS_SCHEMA.CLAIMS
GROUP BY CLAIM_STATE
ORDER BY TOTAL_CLAIMS DESC;
```

```
%%sql -r df_demo_q3
-- Demo Q3: Which region has the maximum number of claims backlog?
SELECT 
    CLAIM_STATE AS REGION,
    COUNT(*) AS BACKLOG_COUNT,
    ROUND(AVG(DATEDIFF(DAY, OPEN_DATE, CURRENT_DATE())), 0) AS AVG_AGE_DAYS
FROM INSURANCE_CLAIM_DB.CLAIMS_SCHEMA.CLAIMS
WHERE CLAIM_STATUS IN ('Open', 'Pending Review', 'Stalled')
  AND CLOSE_DATE IS NULL
GROUP BY CLAIM_STATE
ORDER BY BACKLOG_COUNT DESC;
```

```
%%sql -r df_demo_q4
-- Demo Q4: Which stage has max no. of stalled claims in the past quarter?
SELECT 
    CLAIM_STAGE AS STAGE,
    COUNT(*) AS STALLED_CLAIMS,
    ROUND(AVG(DATEDIFF(DAY, LAST_ACTIVITY_DATE, CURRENT_DATE())), 0) AS AVG_DAYS_STALLED
FROM INSURANCE_CLAIM_DB.CLAIMS_SCHEMA.CLAIMS
WHERE CLAIM_STATUS = 'Stalled'
  AND LAST_ACTIVITY_DATE >= DATEADD(QUARTER, -1, CURRENT_DATE())
GROUP BY CLAIM_STAGE
ORDER BY STALLED_CLAIMS DESC;
```

```
%%sql -r dataframe_3
-- Weather Data Table

CREATE OR REPLACE VIEW INSURANCE_CLAIM_DB.CLAIMS_SCHEMA.WEATHER_FORECAST_90DAY AS

WITH state_city_map AS (
    SELECT 'CA' AS CLAIM_STATE, 'California' AS STATE_NAME, 'los_angeles' AS PROXY_CITY, 34.0522 AS LATITUDE, -118.2437 AS LONGITUDE, '90001' AS ZIP_CODE UNION ALL
    SELECT 'TX', 'Texas', 'houston', 29.7604, -95.3698, '77001' UNION ALL
    SELECT 'NY', 'New York', 'new_york', 40.7128, -74.0060, '10001' UNION ALL
    SELECT 'FL', 'Florida', 'houston', 25.7617, -80.1918, '33101' UNION ALL
    SELECT 'GA', 'Georgia', 'houston', 33.7490, -84.3880, '30301' UNION ALL
    SELECT 'NC', 'North Carolina', 'new_york', 35.2271, -80.8431, '28201' UNION ALL
    SELECT 'PA', 'Pennsylvania', 'new_york', 39.9526, -75.1652, '19101' UNION ALL
    SELECT 'IL', 'Illinois', 'new_york', 41.8781, -87.6298, '60601' UNION ALL
    SELECT 'OH', 'Ohio', 'new_york', 39.9612, -82.9988, '43201' UNION ALL
    SELECT 'MI', 'Michigan', 'new_york', 42.3314, -83.0458, '48201'
),

forecast_latest AS (
    SELECT
        f.*,
        ROW_NUMBER() OVER (PARTITION BY f.CITY_NAME, f.DATE_VALID_STD ORDER BY f.TIME_INIT_UTC DESC) AS RN
    FROM GLOBAL_WEATHER__CLIMATE_DATA_BY_PELMOREX_WEATHER_SOURCE.PWS_BI_SAMPLE.POINT_FORECAST_DAY f
    WHERE f.COUNTRY_CODE = 'US'
      AND f.DATE_VALID_STD BETWEEN CURRENT_DATE() AND DATEADD(DAY, 90, CURRENT_DATE())
),

forecast_data AS (
    SELECT
        scm.CLAIM_STATE,
        scm.STATE_NAME,
        fl.CITY_NAME AS SOURCE_CITY,
        fl.DATE_VALID_STD AS FORECAST_DATE,
        scm.LATITUDE AS LATITUDE_DEG,
        scm.LONGITUDE AS LONGITUDE_DEG,
        scm.ZIP_CODE,
        fl.MIN_TEMPERATURE_AIR_2M_F,
        fl.AVG_TEMPERATURE_AIR_2M_F,
        fl.MAX_TEMPERATURE_AIR_2M_F,
        fl.MIN_HUMIDITY_RELATIVE_2M_PCT,
        fl.AVG_HUMIDITY_RELATIVE_2M_PCT,
        fl.MAX_HUMIDITY_RELATIVE_2M_PCT,
        fl.AVG_PRESSURE_MEAN_SEA_LEVEL_MB,
        fl."__MAX_WIND_SPEED_10M_MPH" AS MAX_WIND_SPEED_MPH,
        fl."__AVG_WIND_SPEED_10M_MPH" AS AVG_WIND_SPEED_MPH,
        fl.AVG_WIND_DIRECTION_10M_DEG AS WIND_DIRECTION_DEG,
        fl.TOT_PRECIPITATION_IN,
        fl.TOT_SNOWFALL_IN,
        fl.AVG_CLOUD_COVER_TOT_PCT,
        fl.PROBABILITY_OF_PRECIPITATION_PCT,
        fl.PROBABILITY_OF_SNOW_PCT,
        'FORECAST' AS DATA_SOURCE,
        CASE
            WHEN fl.TOT_SNOWFALL_IN > 0 AND fl."__MAX_WIND_SPEED_10M_MPH" >= 35 THEN 'Blizzard'
            WHEN fl.TOT_PRECIPITATION_IN >= 2.0 THEN 'Heavy Rain'
            WHEN fl."__MAX_WIND_SPEED_10M_MPH" >= 58 THEN 'Severe Wind'
            WHEN fl."__MAX_WIND_SPEED_10M_MPH" >= 40 THEN 'High Wind'
            WHEN fl.TOT_PRECIPITATION_IN >= 0.5 THEN 'Moderate Rain'
            WHEN fl.TOT_SNOWFALL_IN > 0 THEN 'Snow'
            WHEN fl.TOT_PRECIPITATION_IN > 0 THEN 'Light Rain'
            ELSE 'Clear/Dry'
        END AS WEATHER_CONDITION,
        CASE
            WHEN fl."__MAX_WIND_SPEED_10M_MPH" >= 58 THEN 'High'
            WHEN fl.TOT_PRECIPITATION_IN >= 2.0 THEN 'High'
            WHEN fl."__MAX_WIND_SPEED_10M_MPH" >= 40 OR fl.TOT_SNOWFALL_IN >= 6 THEN 'Medium'
            WHEN fl.TOT_PRECIPITATION_IN >= 0.5 OR fl.TOT_SNOWFALL_IN > 0 THEN 'Low'
            ELSE 'None'
        END AS WEATHER_RISK_LEVEL
    FROM forecast_latest fl
    JOIN state_city_map scm ON fl.CITY_NAME = scm.PROXY_CITY
    WHERE fl.RN = 1
),

date_spine AS (
    SELECT DATEADD(DAY, SEQ4(), CURRENT_DATE()) AS FORECAST_DATE
    FROM TABLE(GENERATOR(ROWCOUNT => 90))
),

climatology_data AS (
    SELECT
        scm.CLAIM_STATE,
        scm.STATE_NAME,
        c.CITY_NAME AS SOURCE_CITY,
        ds.FORECAST_DATE,
        scm.LATITUDE AS LATITUDE_DEG,
        scm.LONGITUDE AS LONGITUDE_DEG,
        scm.ZIP_CODE,
        c.AVG_OF__DAILY_MIN_TEMPERATURE_AIR_2M_F AS MIN_TEMPERATURE_AIR_2M_F,
        c.AVG_OF__DAILY_AVG_TEMPERATURE_AIR_2M_F AS AVG_TEMPERATURE_AIR_2M_F,
        c.AVG_OF__DAILY_MAX_TEMPERATURE_AIR_2M_F AS MAX_TEMPERATURE_AIR_2M_F,
        c.AVG_OF__DAILY_MIN_HUMIDITY_RELATIVE_2M_PCT AS MIN_HUMIDITY_RELATIVE_2M_PCT,
        c.AVG_OF__DAILY_AVG_HUMIDITY_RELATIVE_2M_PCT AS AVG_HUMIDITY_RELATIVE_2M_PCT,
        c.AVG_OF__DAILY_MAX_HUMIDITY_RELATIVE_2M_PCT AS MAX_HUMIDITY_RELATIVE_2M_PCT,
        c.AVG_OF__DAILY_AVG_PRESSURE_MEAN_SEA_LEVEL_MB AS AVG_PRESSURE_MEAN_SEA_LEVEL_MB,
        c.AVG_OF__DAILY_MAX_WIND_SPEED_10M_MPH AS MAX_WIND_SPEED_MPH,
        c.AVG_OF__DAILY_AVG_WIND_SPEED_10M_MPH AS AVG_WIND_SPEED_MPH,
        c.AVG_VEC_OF__DAILY_AVG_VEC_WIND_DIRECTION_10M_DEG AS WIND_DIRECTION_DEG,
        c.AVG_OF__POS_DAILY_TOT_PRECIPITATION_IN AS TOT_PRECIPITATION_IN,
        c.AVG_OF__POS_DAILY_TOT_SNOWFALL_IN AS TOT_SNOWFALL_IN,
        c.AVG_OF__DAILY_AVG_CLOUD_COVER_TOT_PCT AS AVG_CLOUD_COVER_TOT_PCT,
        c.FRQ_OF__POS_DAILY_TOT_PRECIPITATION_IN_PCT AS PROBABILITY_OF_PRECIPITATION_PCT,
        c.FRQ_OF__POS_DAILY_TOT_SNOWFALL_IN_PCT AS PROBABILITY_OF_SNOW_PCT,
        'CLIMATOLOGY' AS DATA_SOURCE,
        CASE
            WHEN c.AVG_OF__POS_DAILY_TOT_SNOWFALL_IN > 0 AND c.AVG_OF__DAILY_MAX_WIND_SPEED_10M_MPH >= 35 THEN 'Blizzard'
            WHEN c.AVG_OF__POS_DAILY_TOT_PRECIPITATION_IN >= 2.0 THEN 'Heavy Rain'
            WHEN c.AVG_OF__DAILY_MAX_WIND_SPEED_10M_MPH >= 40 THEN 'High Wind'
            WHEN c.AVG_OF__POS_DAILY_TOT_PRECIPITATION_IN >= 0.5 THEN 'Moderate Rain'
            WHEN c.AVG_OF__POS_DAILY_TOT_SNOWFALL_IN > 0 THEN 'Snow'
            WHEN c.AVG_OF__POS_DAILY_TOT_PRECIPITATION_IN > 0 THEN 'Light Rain'
            ELSE 'Clear/Dry'
        END AS WEATHER_CONDITION,
        CASE
            WHEN c.AVG_OF__DAILY_MAX_WIND_SPEED_10M_MPH >= 58 THEN 'High'
            WHEN c.AVG_OF__POS_DAILY_TOT_PRECIPITATION_IN >= 2.0 THEN 'High'
            WHEN c.AVG_OF__DAILY_MAX_WIND_SPEED_10M_MPH >= 40 OR c.AVG_OF__POS_DAILY_TOT_SNOWFALL_IN >= 6 THEN 'Medium'
            WHEN c.AVG_OF__POS_DAILY_TOT_PRECIPITATION_IN >= 0.5 OR c.AVG_OF__POS_DAILY_TOT_SNOWFALL_IN > 0 THEN 'Low'
            ELSE 'None'
        END AS WEATHER_RISK_LEVEL
    FROM date_spine ds
    CROSS JOIN state_city_map scm
    JOIN GLOBAL_WEATHER__CLIMATE_DATA_BY_PELMOREX_WEATHER_SOURCE.PWS_BI_SAMPLE.POINT_CLIMATOLOGY_DAY c
        ON c.CITY_NAME = scm.PROXY_CITY
        AND c.DOY_STD = DAYOFYEAR(ds.FORECAST_DATE)
        AND c.COUNTRY_CODE = 'US'
    WHERE NOT EXISTS (
        SELECT 1 FROM forecast_data fd
        WHERE fd.CLAIM_STATE = scm.CLAIM_STATE
          AND fd.FORECAST_DATE = ds.FORECAST_DATE
    )
)

SELECT * FROM forecast_data
UNION ALL
SELECT * FROM climatology_data
ORDER BY CLAIM_STATE, FORECAST_DATE
```
